```python
import pandas as pd
import numpy as np

class DatasetLoader:
    def __init__(self, csv_path: str):
        self.df = pd.read_csv(csv_path)

    @property
    def data(self):
        return self.df.iloc[:, :7].to_numpy()

    @property
    def target(self):
        return np.squeeze(self.df.iloc[:, 7:].to_numpy(), axis=-1)

```

```python
from sklearn.model_selection import train_test_split

dset = DatasetLoader("posture_dataset.csv")

X_train, X_test, y_train, y_test = train_test_split(dset.data, dset.target, test_size=0.75, random_state=42)

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rnd_clf = RandomForestClassifier(n_estimators=100, random_state=42)

rnd_clf.fit(X_train, y_train)

y_pred = rnd_clf.predict(X_test)

print(accuracy_score(y_test, y_pred))
```

```python
import pickle

with open("./model.pkl", "wb") as f:
    pickle.dump(rnd_clf, f)
```
